# Outlier detection

```python
# Z-score 
from scipy import stats
import numpy as np

cols_to_check = ['AGE', 'UHRSWORKT', 'WKSWORK1', 'UHRSWORKLY', 'INCTOT']

# culculate Z-Score
z_scores = stats.zscore(df_encoded[cols_to_check])

# define the threshold
threshold = 3
outliers_mask = (z_scores > threshold).any(axis=1)

# delete the outiler lines
df_cleaned = df_encoded[~outliers_mask]
print(df_cleaned)
```

# Data Preprocessing

### Normalization (for numurical data)

```python
from sklearn.preprocessing import MinMaxScaler
df_final_nor = df_final.copy()
cols_to_normalize = ['AGE', 'UHRSWORKT', 'WKSWORK1', 'UHRSWORKLY', 'INCTOT']
scaler = MinMaxScaler()
df_final_nor[cols_to_normalize] = scaler.fit_transform(df_final[cols_to_normalize])
df_final_nor
```

# Find the hyperparameters 

```python
from sklearn.model_selection import train_test_split

# get X and Y
feature_columns = df_final_nor.loc[:, df_final_nor.columns != 'INCTOT'].columns
X = df_final_nor[feature_columns]
y = df_final_nor["INCTOT"]

# split train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
train_x = X_train.values
train_y = y_train.values
test_x = X_test.values
test_y = y_test.values
```

### Based on MSE, find the optimal hyperparameters (use cross-validaiton)

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

lasso_reg = Lasso()
param_grid_lasso = {'alpha': [0.001, 0.01, 0.1, 1, 10]}

lasso_grid = GridSearchCV(lasso_reg, param_grid_lasso, cv=5)
lasso_grid.fit(X_train, y_train)

print("Lasso Regression optimal hyperparameter: ", lasso_grid.best_params_)
```

```python
from sklearn.svm import SVR
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge

ridge_reg = Ridge()
param_grid_ridge = {'alpha': [0.001, 0.01, 0.1, 1, 2, 10],
                    'solver': ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg']}

ridge_grid = GridSearchCV(ridge_reg, param_grid_ridge, cv=5)
ridge_grid.fit(X_train, y_train)

print("Ridge Regression optimal hyperparameter: ", ridge_grid.best_params_)
```

# Train the models and see the performance metrics of each models

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

lasso_clf = Lasso(alpha=0.001)
ridge_clf = Ridge(alpha=1, solver='sparse_cg')
# svr_clf = SVR(kernel='linear')

# Train the models
lasso_clf.fit(X_train, y_train)
ridge_clf.fit(X_train, y_train)
# svr_clf.fit(X_train, y_train)

# predict the models
lasso_pred = lasso_clf.predict(X_test)
ridge_pred = ridge_clf.predict(X_test)
# svr_pred = svr_clf.predict(X_test)

# calculate MSE
lasso_mse = mean_squared_error(y_test, lasso_pred)
ridge_mse = mean_squared_error(y_test, ridge_pred)
# svr_mse = mean_squared_error(y_test, svr_pred)

print("Lasso MSE:", lasso_mse)
print("Ridge MSE:", ridge_mse)
# print("SVR MSE:", svr_mse)

# Calculate MAE
lasso_mae = mean_absolute_error(y_test, lasso_pred)
ridge_mae = mean_absolute_error(y_test, ridge_pred)
# svr_mae = mean_absolute_error(y_test, svr_pred)

print("Lasso MAE:", lasso_mae)
print("Ridge MAE:", ridge_mae)
# print("SVR MAE:", svr_mae)
```
