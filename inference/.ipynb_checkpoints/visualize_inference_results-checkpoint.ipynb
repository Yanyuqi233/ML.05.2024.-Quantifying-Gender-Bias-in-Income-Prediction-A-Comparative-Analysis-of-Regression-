# Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Data

```python
fp = '/Users/niahodges/Downloads/Spring 2024/Practicum in Data Analytics and Statistics/project/scripts/inference/final_preds.csv'
df = pd.read_csv(fp)
```

```python
df
```

```python
g1_df = df[df['SEX_1'] == 0]
g2_df = df[df['SEX_1'] == 1]

important_cols = ['SEX_1', 'lasso_preds', 'ridge_preds']
g1_df = g1_df[important_cols].reset_index()
g2_df = g2_df[important_cols].reset_index()
df_small = df[important_cols]
df_small['idx'] = df_small.index.tolist()
```

```python
g1_df.shape
```

```python
g2_df.shape
```

# Visualizations

## Lasso Regression

```python
model = 'Lasso Regression'

plt.plot(g1_df.index, g1_df['lasso_preds'], label='Gender 1')
plt.plot(g1_df.index, g2_df['lasso_preds'], label='Gender 2', alpha=0.5)

# Set labels and title
plt.xlabel('Test Cases')
plt.ylabel('Income Prediction')
plt.title(f'Causality Analysis in {model} Model')

# Display legend
plt.legend()

# Show the plot
plt.show()
```

```python
model = 'Lasso Regression'
start = 0
stop = 20

temp_df = df_small[start:stop]

grouped_df = temp_df.groupby(['idx', 'SEX_1'])['lasso_preds'].sum().unstack()

# Plot the grouped DataFrame as a bar chart
grouped_df.plot(kind='bar', stacked=True)

# Set labels and title
plt.xlabel('Test Cases')
plt.ylabel('Income Prediction')
plt.title(f'Causality Analysis in {model} for Cases {start}-{stop}')

# Show the plot
plt.show()
```

```python
model = 'Lasso Regression'
start = 20
stop = 40

temp_df = df_small[start:stop]

grouped_df = temp_df.groupby(['idx', 'SEX_1'])['lasso_preds'].sum().unstack()

# Plot the grouped DataFrame as a bar chart
grouped_df.plot(kind='bar', stacked=True)

# Set labels and title
plt.xlabel('Test Cases')
plt.ylabel('Income Prediction')
plt.title(f'Causality Analysis in {model} for Cases {start}-{stop}')

# Show the plot
plt.show()
```
