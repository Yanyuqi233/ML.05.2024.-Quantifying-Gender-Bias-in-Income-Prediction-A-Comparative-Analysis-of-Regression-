# Data

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.svm import SVR
from util_funcs import process_data, split_data
```

```python
df = pd.read_csv('/Users/niahodges/Downloads/Spring 2024/Practicum in Data Analytics and Statistics/project/temp-data/cps_ALL.csv')
```

```python
df
```

```python
df_temp1 = df.copy()
df_gender = process_data(df_temp1,'gender')
df_temp2 = df.copy()
df_noGender = process_data(df_temp2,'no-gender')
```

```python
X_train_G, X_test_G, y_train_G, y_test_G = split_data(df_gender)
X_train_NG, X_test_NG, y_train_NG, y_test_NG = split_data(df_noGender)
```

# Parameters for gender inclusive models

```python
lasso_reg_G = Lasso()
ridge_reg_G = Ridge()
svr_reg_G = SVR()
```

```python
lasso_params = {
    'alpha': [0.001, 0.01, 0.1, 1, 10],
    'fit_intercept': [True, False],
    'max_iter': [1000]
}
ridge_params = {
    'alpha': [0.001, 0.01, 0.1, 1, 10],
    'solver': ['auto', 'svd', 'cholesky', 'lsqr', 'sparse_cg'],
    'fit_intercept': [True, False],
    'max_iter': [1000]
}
svr_params = {
    'C': [0.001, 0.01, 0.1, 1, 10],
    'kernel': ['linear', 'rbf', 'poly'],
    'epsilon': [0.01, 0.1, 1, 10]
}
```

```python
lasso_grid_G = GridSearchCV(lasso_reg_G, lasso_params, cv=5, verbose=1)
lasso_grid_G.fit(X_train_G, y_train_G)
print(f'Lasso regression: {lasso_grid_G.best_params_}')
```

```python
ridge_grid_G = GridSearchCV(ridge_reg_G, ridge_params, cv=5, verbose=1)
ridge_grid_G.fit(X_train_G, y_train_G)
print(f'Ridge regression: {ridge_grid_G.best_params_}')
```

```python
svr_grid_G = GridSearchCV(svr_reg_G, svr_params, cv=5, verbose=1)
svr_grid_G.fit(X_train_G, y_train_G)
print(f'SVR regression: {svr_grid_G.best_params_}')
```
