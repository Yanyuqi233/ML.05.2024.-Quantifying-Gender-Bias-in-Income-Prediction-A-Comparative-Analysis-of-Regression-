# Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Visualizations

## Employed persons by occupation, sex, and age

```python
file_path = '/Users/niahodges/Downloads/Spring 2024/Practicum in Data Analytics and Statistics/project/temp-data/temp1.xlsx'
file_path = '../Data/statistical data/temp1.xlsx'
df1 = pd.read_excel(file_path, header=0)
```

```python
df1
```

```python
# Group the data by Occupation and Gender, and calculate the average income
grouped_data = df1.groupby(['Year', 'Occupation'])['Avg-Income'].mean().unstack()

# Plot the grouped data as a stacked bar chart
grouped_data.plot(kind='bar', stacked=False)

# Set the labels and title
plt.xlabel('Year')
plt.ylabel('Average Annual Income')
plt.title('Average Annual Income (In Thousands) of Employed Persons By Occupation and Year')

# Show the plot
plt.legend(title='Occupation')
plt.savefig('/Users/niahodges/Downloads/Spring 2024/Practicum in Data Analytics and Statistics/project/figures/employment-statistics/graph1')
plt.show()
```

## Employment status of the civilian noninstitutional population 16 years and over by sex, 1983 to date

```python
file_path = '/Users/niahodges/Downloads/Spring 2024/Practicum in Data Analytics and Statistics/project/temp-data/temp2.xlsx'
df2 = pd.read_excel(file_path, header=0)
```

```python
df2
```

```python
# Group the data by Occupation and Gender, and calculate the average income
grouped_data = df2.groupby(['Year', 'Gender'])['Employed'].mean().unstack()

# Plot the grouped data as a stacked bar chart
grouped_data.plot(kind='line', stacked=False)

# Set the labels and title
plt.xlabel('Year')
plt.ylabel('Number of Employed Persons (In Thousands)')
plt.title('Number of Persons (In Thousands) Employed By Year and Gender')

# Show the plot
plt.legend(title='Gender')
plt.savefig('/Users/niahodges/Downloads/Spring 2024/Practicum in Data Analytics and Statistics/project/figures/employment-statistics/graph2')
plt.show()
```

```python
# Group the data by Occupation and Gender, and calculate the average income
grouped_data = df2.groupby(['Year', 'Gender'])['Unemployed'].mean().unstack()

# Plot the grouped data as a stacked bar chart
grouped_data.plot(kind='line', stacked=False)

# Set the labels and title
plt.xlabel('Year')
plt.ylabel('Number of Unemployed Persons (In Thousands)')
plt.title('Number of Persons (In Thousands) Unemployed By Year and Gender')

# Add vertical lines at specific years
plt.axvline(x=2010, color='red', linestyle='--', label='The Great Recession')
plt.axvline(x=2020, color='red', linestyle='--', label='Covid-19')

# Show the plot
plt.legend(title='Gender')
plt.savefig('/Users/niahodges/Downloads/Spring 2024/Practicum in Data Analytics and Statistics/project/figures/employment-statistics/graph3')
plt.show()
```

```python
temp_df2 = df2[df2['Gender'] == 'Women']
temp_df2
```

```python
temp_df2.columns
```

```python
# Filter the data for a specific year
target_year = 2023
specific_year_data = temp_df2[temp_df2['Year'] == target_year]

# Calculate the sum of in-workforce and not-in-workforce counts
in_workforce_count = specific_year_data['In-Labor-Force'].item()
not_in_workforce_count = specific_year_data['Not-In-Labor-Force'].item()

# Create a pie chart
labels = ['In Workforce', 'Not In Workforce']
sizes = [in_workforce_count, not_in_workforce_count]
colors = ['lightblue', 'lightcoral']
explode = (0.1, 0)  # explode the 1st slice
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Set the title
plt.title(f'Distribution of Women in the Labor Force in {target_year}')

# Show the plot
plt.savefig('/Users/niahodges/Downloads/Spring 2024/Practicum in Data Analytics and Statistics/project/figures/employment-statistics/graph4')
plt.show()
```

## Employment Status by Race and Year

```python
file_path = '/Users/niahodges/Downloads/Spring 2024/Practicum in Data Analytics and Statistics/project/temp-data/temp3.xlsx'
df3 = pd.read_excel(file_path, header=0)
```

```python
df3
```

```python
temp_df3 = df3[df3['Employment-status'] == 'Employed']
temp_df3
```

```python
# Group the data by Occupation and Gender, and calculate the average income
grouped_data = temp_df3.groupby(['Year', 'Race'])['Total'].mean().unstack()

# Plot the grouped data as a stacked bar chart
grouped_data.plot(kind='bar', stacked=False)

# Set the labels and title
plt.xlabel('Year')
plt.ylabel('Number of Employed Persons (In Thousands) by Race')
plt.title('Number of Persons (In Thousands) Employed By Year and Race')

# Show the plot
plt.legend(title='Race')
plt.savefig('/Users/niahodges/Downloads/Spring 2024/Practicum in Data Analytics and Statistics/project/figures/employment-statistics/graph9')
plt.show()
```

```python
temp2_df3 = df3[df3['Employment-status'] == 'Employed']
temp2_df3
```

```python
# Group the data by Occupation and Gender, and calculate the average income
grouped_data = temp2_df3.groupby(['Year', 'Race'])['Total'].mean().unstack()

# Plot the grouped data as a stacked bar chart
grouped_data.plot(kind='bar', stacked=False)

# Set the labels and title
plt.xlabel('Year')
plt.ylabel('Number of Unemployed Persons (In Thousands) by Race')
plt.title('Number of Persons (In Thousands) Unemployed By Year and Race')

# Show the plot
plt.legend(title='Race')
plt.savefig('/Users/niahodges/Downloads/Spring 2024/Practicum in Data Analytics and Statistics/project/figures/employment-statistics/graph10')
plt.show()
```

```python
temp3_df3 = df3[df3['Year']==2023]
temp3_df3
```

```python
# Group the data by Occupation and Gender, and calculate the average income
grouped_data = temp3_df3.groupby(['Race', 'Employment-status'])['Total'].mean().unstack()

# Plot the grouped data as a stacked bar chart
grouped_data.plot(kind='bar', stacked=True)

# Set the labels and title
plt.xlabel('Race')
plt.ylabel('Number of Persons (In Thousands) by Race')
plt.title('Number of Persons (In Thousands) By Race in 2023')

# Show the plot
plt.legend(title='Race')
plt.savefig('/Users/niahodges/Downloads/Spring 2024/Practicum in Data Analytics and Statistics/project/figures/employment-statistics/graph11')
plt.show()
```

```python
race = 'Black'
status = 'Employed'
index = temp3_df3.loc[(temp3_df3['Race'] == race) & (temp3_df3['Employment-status'] == status)].index.item()
```

```python
race = 'Black'
status1 = 'Employed'
status2 = 'Unemployed'
employed_index = temp3_df3.loc[(temp3_df3['Race'] == race) & (temp3_df3['Employment-status'] == status1)].index.item()
unemployed_index = temp3_df3.loc[(temp3_df3['Race'] == race) & (temp3_df3['Employment-status'] == status2)].index.item()


# Calculate the sum of in-workforce and not-in-workforce counts
employment_count = temp3_df3['Total'][employed_index]
unemployment_count = temp3_df3['Total'][unemployed_index]

# Create a pie chart
labels = ['Employed', 'Unemployed']
sizes = [employment_count, unemployment_count]
colors = ['lightblue', 'lightcoral']
explode = (0.1, 0)  # explode the 1st slice
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Set the title
plt.title(f'Distribution of Employed and Unemployed {race} Persons in 2023')

# Show the plot
plt.savefig('/Users/niahodges/Downloads/Spring 2024/Practicum in Data Analytics and Statistics/project/figures/employment-statistics/graph12')
plt.show()
```
