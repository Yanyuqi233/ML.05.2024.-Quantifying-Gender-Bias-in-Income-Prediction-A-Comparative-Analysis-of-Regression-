## Gender inclusive

```python
directory = '/Users/niahodges/Downloads/SPring 2024/Practicum in Data Analytics and Statistics/project/scripts/training/logs/linear models feature weights/gender inclusive'
lassoWeights_fn = 'lasso_weights.txt'
ridgeWeights_fn = 'ridge_weights.txt'
```

```python
lasso_fp = os.path.join(directory, lassoWeights_fn)
ridge_fp = os.path.join(directory, ridgeWeights_fn)
```

```python
file_path = lasso_fp

feature_names = []
weights = []

# Read the file line by line, skipping the first two lines
with open(file_path, 'r') as file:
    line_number = 0
    for line in file:
        # Skip the first two lines (headers)
        if line_number < 2:
            line_number += 1
            continue
        
        # Split the line at the last space
        split_line = line.rsplit(' ', 1)
        
        # Process the feature name (dropping the last character)
        feature_name = split_line[0][:-1]
        
        # Convert the weight to a float number
        weight = float(split_line[1])
        
        # Append to respective lists
        feature_names.append(feature_name)
        weights.append(weight)
        
        line_number += 1
        
lassoFeatures_G_df = pd.DataFrame({'feature':feature_names,
                               'weight':weights})
```

```python
file_path = ridge_fp

feature_names = []
weights = []

# Read the file line by line, skipping the first two lines
with open(file_path, 'r') as file:
    line_number = 0
    for line in file:
        # Skip the first two lines (headers)
        if line_number < 2:
            line_number += 1
            continue
        
        # Split the line at the last space
        split_line = line.rsplit(' ', 1)
        
        # Process the feature name (dropping the last character)
        feature_name = split_line[0][:-1]
        
        # Convert the weight to a float number
        weight = float(split_line[1])
        
        # Append to respective lists
        feature_names.append(feature_name)
        weights.append(weight)
        
        line_number += 1
        
ridgeFeatures_G_df = pd.DataFrame({'feature':feature_names,
                               'weight':weights})
```
